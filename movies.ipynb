# Natural Language Processing

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text = str(text).lower()  #lowercase
  text = re.sub(r'[^a-zA-Z\s]', '', text)  #remove non-alphanumeric/punctuations

  words = text.split()  # tokenization
  words = [word for word in words if word not in stop_words]  #remove stopwords
  words = [lemmatizer.lemmatize(word) for word in words]  #lemmatization/Basic form

  return ' '.join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

## Converting 'tags' to numerical verctor using TF-IDF

```python
# To handle the missing index, it reset the index
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index, index=df['title'])
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features= 50000, ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

## The Actual Prediction- Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
  if title not in indices:
    return "Movie Not Found!"

  idx = indices[title]  # to get the index of the input movie

  # to get the similarity scores & flatten() is to get the score in 1D
  sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  # argsort() will return the sorted index of top 10 movies
  # which are then arranged in assending order
  similar_idx = sim_scores.argsort()[::-1][1:n+1]

  # retrieving the top 10 similar movies based on indexes
  return df['title'].iloc[similar_idx]
```

```python
recommend('Toy Story')
```
